```
import pandas as pd

# Load LAB base results and RESPONDENTI files
df_base_lab = pd.read_excel('LAB results base - aSTRAV.xlsx')
df_respondenti = pd.read_excel('RESPONDENTI.xlsx')

# Get list/Series of valid IDs
valid_ids = df_respondenti['ID_ID']

# Filter LAB base results
filtered_lab = df_base_lab[df_base_lab['ID_ID'].isin(valid_ids)]

print(filtered_lab.head())

filtered_lab
```

```
median_value = filtered_lab['aSTRAV-DELKA'].median()
print("Median of aSTRAV-DELKA (filtered_lab):", median_value)
```
