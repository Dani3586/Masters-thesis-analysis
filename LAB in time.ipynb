# missing 2 females ferritin 

```
from scipy.stats import shapiro, ttest_rel, wilcoxon

# Extract female Ferritin values
female_base = merged_females_ferritin_subset_clean['base_Ferritin']
female_follow = merged_females_ferritin_subset_clean['followup_Ferritin']

# --- Normality test ---
shapiro_base_f = shapiro(female_base)
shapiro_follow_f = shapiro(female_follow)

print("Shapiro-Wilk test for normality:")
print("Base Ferritin:", shapiro_base_f)
print("Follow-up Ferritin:", shapiro_follow_f)

# --- Choose test ---
# If data is normally distributed: paired t-test
# If not: Wilcoxon signed-rank test
if shapiro_base_f.pvalue > 0.05 and shapiro_follow_f.pvalue > 0.05:
    test_f_ferr = ttest_rel(female_base, female_follow)
    print("\nPaired t-test (FEMALES):")
else:
    test_f_ferr = wilcoxon(female_base, female_follow)
    print("\nWilcoxon signed-rank test (FEMALES):")

print(test_f_ferr)
```
